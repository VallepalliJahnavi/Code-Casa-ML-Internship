```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the MovieLens dataset (assuming you have it uploaded to Colab)
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')

# Create a user-item matrix
user_movie_ratings = ratings.pivot_table(index='userId', columns='movieId', values='rating')

# Define a function to recommend movies
def recommend_movies(user_id, num_recommendations=5):
    # Get the user's ratings
    user_ratings = user_movie_ratings.loc[user_id].dropna()

    # Find similar users (using Pearson correlation)
    similar_users = user_movie_ratings.corrwith(user_ratings)

    # Create a DataFrame of similar users and their correlation scores
    similar_users_df = pd.DataFrame({'user_id': similar_users.index, 'correlation': similar_users.values})
    similar_users_df = similar_users_df.dropna()
    similar_users_df = similar_users_df[similar_users_df['user_id'] != user_id]

    # Sort the similar users by correlation score in descending order
    similar_users_df = similar_users_df.sort_values(by='correlation', ascending=False)

    # Get the movies that the user has not rated
    unrated_movies = movies[~movies['movieId'].isin(user_ratings.index)]

    # Merge the unrated movies with the similar users DataFrame
    recommendations = pd.merge(unrated_movies, similar_users_df, left_on='movieId', right_on='user_id')

    # Sort the recommendations by correlation score in descending order
    recommendations = recommendations.sort_values(by='correlation', ascending=False)

    # Get the top N recommended movies
    top_recommendations = recommendations.head(num_recommendations)

    return top_recommendations[['movieId', 'title']]

# Example usage: Recommend movies for user 1
user_id = 1
recommended_movies = recommend_movies(user_id)
print("Recommended Movies for User 1:")
print(recommended_movies)
```
